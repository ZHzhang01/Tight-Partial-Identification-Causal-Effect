```python
import random
import numpy as np
import matplotlib.pyplot as plt


def getArraySubSet(originArray):
    """
    :type originArray:list
    :rtype :listlist
    """
    result = [[]]
    for i in range(len(originArray)):
        for j in range(len(result)): 
            new_set = result[j]+[originArray[i]]
            result.append(new_set)
            #result.append(result[j]+[originArray[i]]) #现有每个子集中添加新元素，作为新子集加入结果集中
    return result
 
# 测试
originArray=[1,2,3,4,5,6,7]
subset = getArraySubSet(originArray)

print(subset)
print('len(subset):', len(subset))



def getArraySubSet_comp(originArray,left, right):
    """
    :type originArray:list
    :rtype :listlist
    """
    result = [[]]
    
    res = 0
    for i in range(len(originArray)):
        for j in range(len(result)): 
            sum1 = 0
            new_set = result[j]+[originArray[i]]
            result.append(new_set) #现有每个子集中添加新元素，作为新子集加入结果集中           
            for it in new_set:
                sum1 += it
            #print(sum1)
            if sum1 >= left and sum1 <= right:
                res = 1
                return res
    return res


#########################################################################################
def test_sample_twoset(Num, pxy, px, px_y_, px_):
    N = Num # number of integers
    pu = uniform_sampling(N)
    
    condition1  = 0
    result = [[]]
    for i in range(len(pu)):
        for j in range(len(result)): 
            sum = 0
            new_set = result[j]+[pu[i]]
            result.append(new_set) #现有每个子集中添加新元素，作为新子集加入结果集中
            #求和
            for it in new_set:
                sum += it
            if sum >= pxy and sum <= px:
                condition1 = 1
                # 找到pu中对应new_set最后元素的索引号
                set_diff = list(set(pu)-set(new_set))
                #只需比较pu与符合条件集合的差（不相交）集合
                condition2 = getArraySubSet_comp(set_diff,px_y_, px_)
                #print(condition2)
                if condition2 == 1:
                    return pu,condition1
                else:
                    condition1 = 0


    
    return pu, condition1
```
